```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import MultipleLocator
```

```python
## -- load in data from curated npz file
## -- we'll load in the case 2 one, but you could use any of the four
## --      since we're only grabbing the raw light curve
load = np.load('./data/broadband_readysave_case2.npz')
t = load['t'] # times in (MJD)
y = load['y'] # relative fluxes
yerr = load['yerr'] # relative flux uncertainties
del load
```

```python
## -- create figure:
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(t, y, ls='None', marker='o', ms=1, c='black')

ax.set_ylabel("Relative Flux", fontsize=11)
ax.set_xlabel("Time (MJD)", fontsize=11)
ax.ticklabel_format(axis="x", style="plain", useOffset=True, useMathText=True, scilimits=(-5,10))
ax.set_ylim(0.992,1.01)
ax.yaxis.set_major_locator(MultipleLocator(0.002))
plt.savefig('./figures/Figure1.png', dpi=600, bbox_inches='tight')
plt.savefig('./figures/Figure1.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
